```python
# Importación de librerías a utilizar 
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np 
```

```python
# Funciones de usuario que emplearé 
def my_range(start, end, step):
    while start <= end:
        yield start
        start += step
```

```python
# Funciones a emplear para los cálculos (Esto es lo que debo modificar en dado caso de que me cambien la función)
def f(t): 
    x = 2.3*(t-np.sin(t)) 
    return x 
def g(t): 
    y = 2.3*(1-np.cos(t)) 
    return y
```

```python
# Creo listas para almacenar los valores
x = list() 
y = list() 
tvalor = list()

# Manejo inputs para controlar el dominio de la función 

limi = float(input("Introduce un valor de inicio para la función"))
limfi = float(input("Introduce un valor de fin para la función"))
paso = float(input("Introduce el valor del paso para cada valor"))

# Creo la función que recorra el dominio de la función y lo almacene en las listas creadas con anterioridad

for t in my_range(limi, limfi, paso):  
    tvalor.append(t) 


for valor in tvalor:
    x.append(f(valor))
    y.append(g(valor))

# Creo una matriz con numpy para después convertirla a un dataframe

valores = np.array([tvalor,
                    x,
                    y])

# La transpongo porque de lo contrario guardaría los valores de las funciones y las funciones mismas como series o datos

valores= np.transpose(valores)



# Por estética les pongo el nombre a sus columnas 
df = pd.DataFrame(valores, columns=["t", "f(t)", "g(t)"])
df
    
```

```python
# Gráfico los valores 

df.plot(x="f(t)", y="g(t)", grid=True, color="black", figsize=(15,8), 
       xlabel="f(t)", ylabel="g(t)", title="Movimiento de las funciones f(t) y g(t) a lo largo de t")

```
